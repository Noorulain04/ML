# My First TensorFlow Program

## Add Necessary Imports

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## Define and compile the neural network

```python
#Define a simplest Neural Network with one layer and one Neuron
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
#Compile the Neural Network using an Optimizer and a loss function
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## Provide the Data to the Neural Network to train

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

## Train the model on the data

```python
model.fit(xs,ys,epochs= 500)
```

## Use the trained Model to Predict unknown Data

```python
print(model.predict([10.0]))
```
